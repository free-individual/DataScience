```python
import pandas as pd
import numpy as np
```

```python
experience = np.arange(0,11).reshape(-1, 1)
```

```python
experience
```

```python
salary_noise = np.random.normal(0, 5000, experience.shape)
```

```python
salary = 40000 + 5000 * experience + salary_noise
```

```python
print(salary)
```

```python
df = pd.DataFrame({'Experience': experience.flatten(), 'Salary': salary.flatten()})
```

```python
df
```

```python
import seaborn as sns
```

```python
sns.scatterplot(data=df, y='Salary', x='Experience')
```

```python
X = df[['Experience']]
y = df[['Salary']]
from sklearn.linear_model import LinearRegression
```

```python
import matplotlib.pyplot as plt
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
ypred = model.predict(X)
```

```python
plt.scatter(data = df, x='Experience', y='Salary', label='Acutal data')
plt.plot(df['Experience'], ypred, color='red', linewidth=3, label='Model\'s predictions')
plt.legend()
```
