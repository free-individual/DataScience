# Journey to becoming a data scientist.
## 1. Gaining insights from the iris dataset.

### 1.1 Loading the data and setting up the base.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
iris['target'] = load_iris().target
iris.tail()
```

```python
iris.columns = ['s_len', 's_wid', 'p_len', 'p_wid','target']
```

### 1.2 Plotting the data

```python
plt.scatter(iris.p_len, iris.s_len)
plt.xlabel("Petal Length")
plt.ylabel("Sepal Length")
plt.title("Correlation Between Petal width and Sepal Length")
```

```python
plt.scatter(iris.p_wid, iris.s_wid)
plt.xlabel("Petal width")
plt.ylabel("Sepal width")
```

### 1.3 Plotting the data using seaborn to gain more detailed insights.

```python
import seaborn as sns
```

```python
sns.scatterplot(x=iris.p_wid, y=iris.s_wid, hue=iris.target)
```

```python
sns.scatterplot(x=iris.p_len, y=iris.s_len, hue=iris.target)
```

```python
sns.scatterplot(x=iris.p_len, y=iris.s_wid, hue=iris.target)
```

```python
sns.scatterplot(x=iris.s_len, y=iris.p_wid, hue=iris.target)
```

```python
sns.pairplot(iris, hue='target')
```
